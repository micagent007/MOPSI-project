```python
import numpy as np
from numpy.linalg import eig
from numpy.linalg import norm
import matplotlib.pyplot as plt
```

```python
def dimer(u, delta, K, eps, GV, HV):
    n = len(u)
    for k in range(K):
        hu = HV(u)
        va, ve = eig(hu)
        idx = np.argsort(va) 
        va = va[idx]
        ve = ve[:,idx]
        if va[0] < 0 and va[1] > 0 and norm(GV(u)) < eps:
            print("OK")
            return u
        elif va[0] > 0:
            print(va[0])
            return None
        else:
            v1 = np.array([ve[0]])
            u = u - delta * (np.identity(n) - 2*np.transpose(v1).dot(v1)).dot(GV(u))
            print(u)
            print("")
    return None
```

```python
delta = 0.001
K = 1000
eps = 0.01
```
